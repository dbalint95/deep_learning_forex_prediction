Notebook for the preprocessing of the exchange rate data

Saves the preprocessed data in a .csv file

```
import os
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
```

```
# Iterate on .csv files in the given directory, read them into dataframes
directory = '/content/drive/My Drive/deeplearning'
data_df = pd.DataFrame()

def change_time_to_date(series_element):
    date = series_element.split(' ')[0]
    date = str(date[:4]) + "-" + str(date[4:6]) + "-" + str(date[6:])
    return date

for filename in os.listdir(directory):
    if filename.endswith(".csv"):
        df = pd.read_csv(os.path.join(directory, filename), header=None, names=['Time', 'BidQuote', 'AskQuote', 'Volume'])
        df = df.drop('Volume', axis=1)
        df['Date'] = df['Time'].apply(func=change_time_to_date)
        df = df.drop('Time', axis=1)
        df = df[['Date', 'BidQuote', 'AskQuote']]
        data_df = pd.concat([data_df, df])
data_df
```

```
print(data_df.all())
data_df = data_df.groupby(['Date']).mean()
data_df.to_csv("forex_data.csv")
```
